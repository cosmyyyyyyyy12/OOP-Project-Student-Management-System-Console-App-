```python
class Student:
  def __init__(self,name,rollno,marks):
    self.name = name
    self.rollno = rollno
    self.marks = marks

  def calculate_grade(self):
        if self.marks >= 90:
            return "A"
        elif self.marks >= 75:
            return "B"
        elif self.marks >= 60:
            return "C"
        else:
            return "Fail"
```

```python
class Student_Manager:
  def __init__(self):
    self.students = []

  def add_students(self):
    name = input("enter student name:")
    rollno = input("enter student rollno:")
    marks = int(input("enter student marks:"))
    self.students.append(Student(name,rollno,marks))
    print("student added to database")

  def display(self):
    if(len(self.students) == 0):
      print("database is empty")
    for student in self.students:
      print(f"name:{student.name}, rollno:{student.rollno}, marks:{student.marks}, grade:{student.calculate_grade()}")

  def search(self):
    n = input("enter student rollno to search:")
    for student in self.students:
      if student.rollno == n:
        print(f"name:{student.name}, rollno:{student.rollno}, marks:{student.marks}")
        return
    print("student not found")
```

```python
def main():
    manager = Student_Manager()

    while True:
        print("\n1. Add")
        print("2. Display")
        print("3. Search")
        print("4. Exit")

        ch = int(input("Choice: "))

        if ch == 1:
            manager.add_students()
        elif ch == 2:
            manager.display()
        elif ch == 3:
            manager.search()
        elif ch == 4:
            break
        else:
            print("Invalid")
main()
```
